```python
import pandas as pd
import re

# Load the dataset
df = pd.read_csv("employee_master.csv")

# Function to clean and split names
def process_name(name):
    # Remove titles
    name = re.sub(r'^(Dr\.|Mr\.|Ms\.)\s*', '', name, flags=re.IGNORECASE).strip()

    parts = name.split()

    first_name = parts[0] if len(parts) > 0 else ''
    middle_initial = ''
    last_name = ''

    if len(parts) == 2:
        last_name = parts[1]
    elif len(parts) >= 3:
        middle_initial = parts[1][0]  # take only initial
        last_name = parts[-1]

    return pd.Series([first_name, middle_initial, last_name])

# Apply name processing
df[['First_Name', 'Middle_Initial', 'Last_Name']] = df['Employee_Name'].apply(process_name)

# Generate base username: first initial + lastname
df['Username'] = (
    df['First_Name'].str[0].str.lower() +
    df['Last_Name'].str.lower()
)

# Handle duplicate usernames
username_counts = {}
duplicates_handled = 0

final_usernames = []

for uname in df['Username']:
    if uname in username_counts:
        username_counts[uname] += 1
        new_uname = f"{uname}{username_counts[uname]}"
        duplicates_handled += 1
    else:
        username_counts[uname] = 1
        new_uname = uname
    final_usernames.append(new_uname)

df['Username'] = final_usernames

# Display results
print(df.head(8))
print("\nTotal duplicate usernames handled:", duplicates_handled)
```

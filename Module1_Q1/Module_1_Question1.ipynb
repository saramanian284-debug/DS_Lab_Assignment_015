```python
import pandas as pd

# Load dataset
df = pd.read_csv("retail_sales.csv")

# Copy for processing
df_clean = df.copy()

# Create column to track method
df_clean['Imputation_Method'] = None

# Step 1: Calculate medians
regional_median = df_clean.groupby(['ProductCategory', 'Region'])['SalesAmount'].median()
category_median = df_clean.groupby('ProductCategory')['SalesAmount'].median()

# Step 2: Fill missing values
for i, row in df_clean.iterrows():
    if pd.isna(row['SalesAmount']):
        category = row['ProductCategory']
        region = row['Region']

        # Try regional median
        reg_med = regional_median.get((category, region))

        if pd.notna(reg_med):
            df_clean.at[i, 'SalesAmount'] = reg_med
            df_clean.at[i, 'Imputation_Method'] = 'Regional_Median'
        else:
            # Fallback to category median
            cat_med = category_median.get(category)
            df_clean.at[i, 'SalesAmount'] = cat_med
            df_clean.at[i, 'Imputation_Method'] = 'Category_Median'

# Step 3: Save cleaned file
df_clean.to_csv("cleaned_retail_sales.csv", index=False)

# Step 4: Count methods used
method_counts = df_clean['Imputation_Method'].value_counts()

print("Imputation Method Counts:")
print(method_counts)

print("\nPreview of cleaned data:")
print(df_clean.head())
```
